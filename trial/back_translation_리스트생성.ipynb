```python
import pandas as pd
import numpy as np
import re
import csv
import time
import os
import requests

from langdetect import detect
from langdetect.lang_detect_exception import LangDetectException

from selenium import webdriver

from selenium.webdriver.support.ui import WebDriverWait
from bs4 import BeautifulSoup as bs

from selenium.webdriver.common.keys import Keys
```

```python
data = pd.read_csv('train.csv')
data = data.dropna(axis=0)
data = data.drop_duplicates(['문장'], keep='first')
text = data['문장']
label = data['태그']
```

```python
browser = webdriver.Chrome('./chromedriver.exe')

url = 'https://papago.naver.com/'
browser.get(url)
```

```python
text = data['문장']

eng_list = []
for txt in text:
    elem = browser.find_element_by_id("txtSource").clear()
    elem = browser.find_element_by_id("txtSource")
    elem.click()
    #time.wait(3)
    elem.send_keys(txt)
    elem.send_keys(Keys.ENTER)  

    soup = bs(browser.page_source,'html.parser')
    time.sleep(3)

    for i in range(0,1):
        eng = soup.find_all("div",attrs={"id":"txtTarget"})
        print(eng[0].get_text())
        eng_list.append(eng[0].get_text())
```

```python
kor_list = []
for txt in eng_list:
    elem = browser.find_element_by_id("txtSource").clear()
    elem = browser.find_element_by_id("txtSource")
    elem.click()
    #time.wait(3)
    elem.send_keys(txt)
    elem.send_keys(Keys.ENTER)  

    soup = bs(browser.page_source,'html.parser')
    time.sleep(3)

    for i in range(0,1):
        kor = soup.find_all("div",attrs={"id":"txtTarget"})
        print(kor[0].get_text())
        kor_list.append(kor[0].get_text())
```
